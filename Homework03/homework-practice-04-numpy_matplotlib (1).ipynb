```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Создайте график x^2*sin(1/x^2)+x на интервале [-1,1], используя 250 точек. Не забудьте пометить оси!
x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$x^2\\sin(1/x^2)+x$')
plt.show()

# 2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы функций 1/(1+x^2) и 1/x^2 на интервале [5,25].
x = np.linspace(5, 25, 1000)
y1 = 1 / (1 + x**2)
y2 = 1 / x**2
rel_diff = np.abs(1 - y1/y2)

plt.figure(figsize=(8, 6))
plt.semilogy(x, rel_diff)
plt.xlabel('x')
plt.ylabel('Relative Difference')
plt.title('Relative Difference of 1/(1+x^2) and 1/x^2')
plt.show()

# 3. Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.
from scipy.special import jn, yn, kn

x = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.plot(x, jn(2, x), label='Bessel J_2(x)')
plt.plot(x, yn(2, x), label='Bessel Y_2(x)')
plt.plot(x, kn(2, x), label='Bessel K_2(x)')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Bessel Functions')
plt.legend()
plt.show()

# 4. Создайте цветовой график sin(x)sin(y) на интервале [-π,π]×[-π,π].
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=20, cmap='RdGy')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('$\sin(x)\sin(y)$')
plt.show()
```

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(1032206778)

# Generate a random integer from 0 to 9
integral_index = np.random.randint(10)
print(f"Selected integral index: {integral_index}")

# Define the integrals
integrals = [
    lambda x: (x**2 + 3) / (x - 2),
    lambda x: (x + 1) / (x**3 - x**2),
    lambda x: np.exp(1 / x**2) / x**3,
    lambda x: np.cos(np.log(x)) / x,
    lambda x: 1 / (x * (1 + np.log(x)**2)),
    lambda x: np.cos(x)**3,
    lambda x: np.cosh(3 * x)**2,
    lambda x: 1 / (x**2 - 2 * x - 8),
    lambda x: 1 / np.sqrt(2 + 3 * x - 2 * x**2),
    lambda x: (2 * x - 1) / (2 * x + 1)
]

# Define the integration limits
limits = [(3, 4), (-2, -1), (1, 2), (1, np.e), (1, np.e), (0, np.pi / 2), (0, 1 / 3), (2, 3), (3 / 4, 2), (0, 2)]

# Define the analytical solutions
analytical_solutions = [
    np.log(3),
    -1 / 2,
    np.exp(1) - np.exp(1 / 4),
    np.sin(1),
    np.log(np.e),
    2 / 3,
    1 / 9 * (3 * np.sqrt(3) * np.sinh(1) + 9 * np.cosh(1) - 10),
    1 / 4,
    np.pi / 3,
    np.log(3)
]

# Select the integral and limits
integral = integrals[integral_index]
a, b = limits[integral_index]
analytical_solution = analytical_solutions[integral_index]

print(f"Integrating f(x) = {integral.__name__}(x) from {a} to {b}")
print(f"Analytical solution: {analytical_solution}")


def rectangle_method(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    return dx * np.sum(f(x[:-1]))


def trapezoidal_method(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    return dx * (0.5 * f(a) + 0.5 * f(b) + np.sum(f(x[1:-1])))


def simpson_method(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    dx = (b - a) / n
    even_sum = np.sum(f(x[2:-1:2]))
    odd_sum = np.sum(f(x[1:-1:2]))
    return dx / 3 * (f(a) + f(b) + 2 * even_sum + 4 * odd_sum)


n = 10
tolerance = 1e-5
error = 1

while error > tolerance:
    rectangle_integral = rectangle_method(integral, a, b, n)
    trapezoidal_integral = trapezoidal_method(integral, a, b, n)
    simpson_integral = simpson_method(integral, a, b, n)

    error = max(abs(rectangle_integral - analytical_solution),
                abs(trapezoidal_integral - analytical_solution),
                abs(simpson_integral - analytical_solution))

    print(f"n = {n}")
    print(f"Rectangle method: {rectangle_integral}")
    print(f"Trapezoidal method: {trapezoidal_integral}")
    print(f"Simpson method: {simpson_integral}")
    print(f"Error: {error}")

    n *= 10
```
