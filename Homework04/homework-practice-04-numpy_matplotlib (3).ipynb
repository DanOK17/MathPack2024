```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Создайте график x^2*sin(1/x^2)+x на интервале [-1,1], используя 250 точек. Не забудьте пометить оси!
x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$x^2\\sin(1/x^2)+x$')
plt.show()

# 2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы функций 1/(1+x^2) и 1/x^2 на интервале [5,25].
x = np.linspace(5, 25, 1000)
y1 = 1 / (1 + x**2)
y2 = 1 / x**2
rel_diff = np.abs(1 - y1/y2)

plt.figure(figsize=(8, 6))
plt.semilogy(x, rel_diff)
plt.xlabel('x')
plt.ylabel('Relative Difference')
plt.title('Relative Difference of 1/(1+x^2) and 1/x^2')
plt.show()

# 3. Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.
from scipy.special import jn, yn, kn

x = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.plot(x, jn(2, x), label='Bessel J_2(x)')
plt.plot(x, yn(2, x), label='Bessel Y_2(x)')
plt.plot(x, kn(2, x), label='Bessel K_2(x)')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Bessel Functions')
plt.legend()
plt.show()

# 4. Создайте цветовой график sin(x)sin(y) на интервале [-π,π]×[-π,π].
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=20, cmap='RdGy')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('$\sin(x)\sin(y)$')
plt.show()
```
